10. How can you tell the difference between break and continue?

```python
                                                        ###  Answer  ###                 
'''

Break => Loops can be exited once a specified condition is met.
Continue => A single loop iteration can be exited by using the continue statement, which jumps to the next loop iteration

'''

for a in range(0,5):
  if a == 2:
    break
  print(f'Iteration: {a}')

print(" ")

for a in range(0,5):
  if a == 2:
    continue
  print(f'Iteration: {a}')
```

11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
                                                        ###  Answer  ###                 
for a in range(5):        # This will start from 0.
  print(f'a : {a}')

print(" ")

for a in range(0,5):        # This will start from left value i.e. min and max range value is given.
  print(f'a : {a}')

print(" ")

for a in range(3,5):        # This will start from left value i.e. min and max range value is given.
  print(f'a : {a}')

print(" ")

for a in range(0,5,2):        # This will start from left value i.e. min and max range value is given. And with some increment value.
  print(f'a : {a}')
```
