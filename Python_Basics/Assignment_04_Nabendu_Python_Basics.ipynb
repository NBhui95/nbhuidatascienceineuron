13. What's the difference between tuples and lists?

```python
                                                        ###  Answer  ###                 
'''

Tuples =>  Tuples are immutable; they cannot be changed at all. Tuples are written using parentheses ().
Lists => Lists are mutable - they can have values added, removed, or changed. lists use the square brackets [].

'''

bacon5 = (3.14, 'cat', 11, 'cat', True)
print(bacon5)
#bacon5[2]="rat"               # TypeError: 'tuple' object does not support item assignment

bacon6 = [3.14, 'cat', 11, 'cat', True]
print(bacon6)
bacon6[2]="rat"
print(bacon6)
```

17. How do you distinguish between copy.copy() and copy.deepcopy()?

```python
                                                        ###  Answer  ###                 
'''

copy.copy() => copy.copy() function will do a shallow copy of a list.

copy.deepcopy() => copy.deepcopy() function will do a deep copy of a list. Only copy.deepcopy() will duplicate any lists inside the list.

'''

# shallow copy

a = [[1, 2], [3, 4]]
#b = a[:]   
import copy
b = copy.copy(a)                 
b[1].append(5)            # Edit the first element (i.e. [1, 2])

print(b)
print(a)                  # It affects the original list.

print(" ")
# deep copy

a = [[1, 2], [3, 4]]
import copy
b = copy.deepcopy(a)               
b[1].append(5)            # Edit the first element (i.e. [1, 2])

print(b)
print(a)                  # It does not affect the original list.
```
