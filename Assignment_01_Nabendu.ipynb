4. What is an expression made up of? What do all expressions do?

```python
                                                        ###  Answer  ###                 
'''

Expression => Combination of Operands and Operators.

Expressions are generally used to express some Arithmetic Operations, Realational Operations, Logical Operations, Bitwise Expression etc.

'''

# Arithmetic Operations (+ - * / // % **)
a = 5
b = 3

addition = a + b
subtraction = a - b
product = a * b
division = a / b
power = a**b

print("Sum of a and b is: ", addition)
print("Difference between a and b is: ", subtraction)
print("Product of a and b is: ", product)
print("Division of a and b is: ", division)
print("a to the power b is: ", power)

# Realational Operations (> < >= <= ==)
p = 10
q = 4
print("p and q are equal => ",p==q)

# Logical Operations (and or not)
x = True
y = False
m = 0
n = 1
print("The result of x and y is: ", x and y)
print("The result of m or n is: ", m or n)

# Bitwise Expression
a = 100
l_shift = a << 1
r_shift = a >> 1

print("One Right shift of a is : ", r_shift)
print("One Left shift of a is : ", l_shift)
```

9. What three functions can be used to get the integer, floating-point number, or string version of a value?

```python
                                                        ###  Answer  ###                 
'''

The int() , float() , and str( ) functions will be used to get the integer, floating-point number, or string version of a value.


'''

a=11
b=11.22
c='Nabendu'

print(f"{a} is {type(a)} data_type.")
print(f"{b} is {type(b)} data_type.")
print(f"{c} is {type(c)} data_type.")
```

10. Why does this expression cause an error? How can you fix it?

 'I have eaten ' + 99 + ' burritos.'

```python
                                                        ###  Answer  ###                 
'''
'I have eaten ' + 99 + ' burritos.'

It will throw following error:

TypeError: can only concatenate str (not "int") to str


We can fix it through type casting (i.e. 'integer to string' for this case only).

'''

print('I have eaten ' + str(99) + ' burritos.')
```
