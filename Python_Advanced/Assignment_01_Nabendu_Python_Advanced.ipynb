Q1. What is the purpose of Python's OOP?

```python
                                                        ###  Answer  ###                 
'''

Object-oriented programming is a programming paradigm that provides a means of structuring programs so that properties and behaviors are bundled into individual objects.

Actually, main purpose is to use :
1.  Class
2.  Objects
3.  Polymorphism
4.  Encapsulation
5.  Inheritance
6.  Data Abstraction

'''
# Class Attribute

class clg_student():
  name=""
  age=0
  dept=""

# Object Creation

stud1=clg_student()
stud1.name="Nabendu"
stud1.age=27
stud1.dept="CSE"

print(f"Student Details => Name : {stud1.name}, Age : {stud1.age} and Department : {stud1.dept}")
```

Q2. Where does an inheritance search look for an attribute?

```python
                                                        ###  Answer  ###                 
'''

An inheritance search looks for an attribute first in the instance object, then in the class the instance was created from, then in all higher superclasses, progressing from left to right.
This search will stop at the first place the attribute will be found.

'''
```

Q3. How do you distinguish between a class object and an instance object?

```python
                                                        ###  Answer  ###                 
'''

Class Object => When we create a class in python then a class object is created so whenever python finds a class statement in the whole program then it creates a class object and assigns a name to that object i.e. class name.
Instance Object => When we call a class, it creates an instance object of that class from which the object has been created.

'''
# Instant Object Creation in Class

class clg_student():
  name=""
  age=0
  dept=""
  def __init__(self, name, age, dept):     # Instant Variable Declaration
    self.name = name
    self.age = age
    self.dept = dept

stud2 = clg_student("Nabendu", 27, "CSE")    # Instance of Class
print(f"Student Name : {stud2.name}")
print(f"Student Age : {stud2.age}")
```

Q8. How would you define the superclasses of a class?

```python
                                                        ###  Answer  ###                 
'''

The class from which the subclass is derived is called a superclass (also a parent class).

'''

class clg_student():
  def __init__(self, id, name, dept):    
    self.id = id 
    self.name = name
    self.dept = dept

class department(clg_student):                      # Defining Superclass
  def __init__(self, id, name, dept, subject): 
    super().__init__(id, name, dept)    
    self.subject = subject


student3 = department(100, "Hrithik", "CSE", "Data Science")
print(f"Student Id : {student3.id}")
print(f"Student Name : {student3.name}")
print(f"Student Department : {student3.dept}")
print(f"Student Subject : {student3.subject}")
```
